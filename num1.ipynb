```python
# задание 1
def function1(s):
  start_index = 0
  end_index = 0
  max_length = 0

  for char in s:
    for i in range(start_index, end_index):
      if char == s[i]:
        start_index = i + 1
        break
      
    end_index += 1
    max_length = max(max_length, (end_index - start_index))

  print(max_length)
  
  
s = 'aabchbad'
function1(s)


# тесты 
# тест 1 показывает, как алгоритм работает с постоянно повторяющимися символами
# тест 2 покрывает все возможные входные данные, 'a' разравает алфавит
test1 = 'a'*1000 + 'b'*1000
function1(test1)

test2 = ('aa' + 'bcdefghijklm' + 'aa' + 'nopqrstuvwxyz')*1000
function1(test2)

# сложность = N*logN
# максимально возможная длина последовательности = количеству уникальных символов, это максимальное количество итераций вложенного цикла
```
